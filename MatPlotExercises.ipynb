```python
plt.plot([10,20,30],[1,2,3])
plt.title('My first plot',{'color':'orange'}, loc='center', pad=10.0, size=40)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.grid(b='True', axis='both', which='major', color='green')
plt.show()
```

```python
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color='orange',label='squares')
plt.plot(t**3,t,color='green',label='cubes')
plt.legend()
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2,t,color='orange')
plt.plot(t**3,t,color='green')
plt.legend(['squares','cubes'])
plt.show()
```

```python
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color='orange')
line2,=plt.plot(t**3,t,color='green')
plt.legend([line1,line2],['squares','cubes'])
plt.show()
```

```python
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color='orange')
line2,=plt.plot(t**3,t,color='green')
plt.legend([line1,line2],['squares','cubes'],loc=10, ncol=2)
plt.show()
```

```python
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color='orange')
line2,=plt.plot(t**3,t,color='green')
plt.xlim(0,100)
plt.ylim(0, 5)
plt.legend([line1,line2],['squares','cubes'],loc=(20/100,2/5), ncol=2)
plt.show()
```

```python
t=np.array([1,2,3,4])
l1,=plt.plot(t**2,t,color='orange')
l2,=plt.plot(t**3,t,color='green')
plt.grid()
plt.legend([l1,l2],['squares','cubes'],loc='lower left',bbox_to_anchor=(0.5,0.5,0.5,0.5), ncol=2)
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2,t,color='orange')
plt.plot(t**3,t,color='green')
plt.xlim(0,100)
plt.ylim(0, 5)
plt.legend(loc='lower right', fontsize=20, facecolor='purple',title='Lines')
plt.show()
```
